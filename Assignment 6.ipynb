```python
import numpy as np #import for use of pi
import matplotlib.pyplot as plt #import for plotting
from scipy import integrate #import for use of integration
   
def normal_density(mean, variance, x):#define function - GREG'S CODE
    fX = 1 / np.sqrt(2 * np.pi * variance)
    exponent = -((x - mean) ** 2) / (2 * variance)
    return fX * np.exp(exponent)

def plot_normal_density(mean, variance, x_range): #define function for plot 
   
    x_values = np.linspace(mean - 3 * np.sqrt(variance), mean + 3 * np.sqrt(variance), 1000)
    y_values = normal_density(mean, variance, x_values) #define y values; follows 0-1000 due to x_value inclusion

    plt.plot(x_values, y_values, label=f'mean={mean}, variance={variance}') #regular code for plotting; mean and variance 
    plt.xlabel('x')
    plt.ylabel('fX(x)')
    plt.title('Normal Density Function')
    plt.legend()
    plt.show()

def numerical_integration(mean, variance, a, b): #define function for integration 
    
    integration = lambda x: normal_density(mean, variance, x) #define integration 
    result, _ = integrate.quad(integration, a, b) #define results
    return result

mean_height = 171 #given height
variance_height = 7.1**2 #given variance_height

plot_normal_density(mean_height, variance_height, (171, 190)) #given values plotted

# probability that a randomly chosen man has a height between 162cm and 190cm
probability = numerical_integration(mean_height, variance_height, 162, 190)
print(f"The probability that a randomly chosen man has a height between 162cm and 190cm is: {probability:.54f}") #statement for probability following graph
```
