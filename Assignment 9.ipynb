```python
# Imports as instructed by Greg

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# a) (i) Defining symbols: : In SymPy, you must explicitly define the symbols you want to use:

x, y = sp.symbols('x y')

# (ii) Differentiation To differentiate a function with respect to a variable:
# Define function f(x, y) = e^x * sin(y) + y^3

f = sp.exp(x) * sp.sin(y) + y**3

# Partial derivatives with respect to x (df/dx) and with respect to y (df/dy):

df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print("Partial derivative of f with respect to x:", df_dx)
print("Partial derivative of f with respect to y:", df_dy)

# (b) Define the function g(x, y) = x^2 * y + x * y^2

g = x**2 * y + x * y**2

# Find the gradient vector and compute the magnitude at point (1, -1)

gradient = [sp.diff(g, i) for i in (x, y)]
gradient_vec = [grad.subs({x: 1, y: -1}).evalf() for grad in gradient]
gradient_magnitude = np.sqrt(sum(float(vec)**2 for vec in gradient_vec))

print("Gradient vector:", gradient_vec)
print("Magnitude of gradient at (1, -1):", gradient_magnitude)

# (c) Define the function h(x, y) = ln(x^2 + y^2)

h = sp.ln(x**2 + y**2)

# Compute second partial derivatives
d2h_dx2 = sp.diff(h, x, x)
d2h_dy2 = sp.diff(h, y, y)
d2h_dxdy = sp.diff(h, x, y)
d2h_dydx = sp.diff(h,x,y)

print("Second partial derivative of h with respect to x:", d2h_dx2)
print("Second partial derivative of h with respect to y:", d2h_dy2)
print("Second partial derivative of h with respect to x and y:", d2h_dxdy)
print("Second partial derivative of h with respect to y and x:", d2h_dydx)

#Symmetry: The mixed second partial derivatives equate to the same value and therefore, d2h/dydx and d2h/dxdy are equal.

# (d) Create a contour plot of j(x, y) = x^3 - 3xy + y^3
# necessary variables already defined (x,y)

j = x**3 - 3*x*y + y**3

j_func = sp.lambdify((x, y), j, 'numpy')
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()

# (e) Create another contour plot of a function of your choice

# 1. f(x, y) = (sin(x))^2 + (cos(y))^3

f = sp.sin(x)**2 + sp.cos(y)**3

f_func = sp.lambdify((x, y), f, 'numpy')
Z = f_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='turbo')
plt.colorbar()
plt.title('Contour plot of $f(x, y) = sin(x)^2 + cos(y)^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()

#I decided to do multiple below this because I thought the output was interesting when playing with the trig functions

# 2. n(x, y) = (sin(x)) + (cos(x))

n = sp.sin(x) + sp.cos(x)

n_func = sp.lambdify((x, y), n, 'numpy')

Z = n_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='rainbow')
plt.colorbar()
plt.title('Contour plot of $n(x, y) = (sin(x)) + cos(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()

# 3. p(x, y) = (cos(x))^2 + (tan(y))^3

p = sp.cos(x)**2 + sp.cos(y)**3

p_func = sp.lambdify((x, y), p, 'numpy')

Z = p_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='cividis')
plt.colorbar()
plt.title('Contour plot of $n(x, y) = (cos(x))^2 + cos(y)^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()

# 3. p(x, y) = (sin(x))^2 + (sin(y))^3

m = sp.sin(x)**2 + sp.sin(y)**3

m_func = sp.lambdify((x, y), m, 'numpy')

Z = m_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='plasma')
plt.colorbar()
plt.title('Contour plot of $n(x, y) = (sin(x))^2 + sin(y)^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
