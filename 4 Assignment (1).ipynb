```python
import matplotlib.pyplot as plt
import numpy as np
f_1 = lambda x: x**2
f_2 = lambda x:((-3)*x**3)+10*x**2+(-5*x)+(-3)
f_3 = lambda x: x**x
f_4 = lambda x: np.abs(x)

def gradient_descent(f, learning_rate, initial_point):

    def derivative(f, base_point): #estimate the derivation of the function f at base_point
        return (f(base_point+10**(-5))-f(base_point))/(10**(-5))
    
    x_coords=[initial_point] #This list is where you will store the x_n's
    y_coords=[f(initial_point)] #This list is where you will store the y_n's
    
    while abs(derivative(f,x_coords[-1]))>0.0000001: #while loop as amount of times needed to run is unknown
        new_x = x_coords[-1] - learning_rate * derivative(f, x_coords[-1]) #define new_x
        x_coords.append(new_x) #x_coords and add new_x
        y_coords.append(f(new_x)) #y_coords and add new_y with respect to x
    
    plot_range=np.linspace(min(x_coords)-0.5, #a nice ploy range
                            max(x_coords)+0.5, 10000) #to make graph look good
    
    function_range=[f(i) for i in plot_range]
    plt.plot(plot_range, function_range) #this plots the function f(x)
    plt.plot(x_coords, y_coords) #this will plot the sequence of points x_n, f(x_n)
    plt.scatter(x_coords, y_coords, color= "lightblue")
    plt.show ()
    
    return ('%.3f' %round(x_coords[-1],3), '%.3f' %round(y_coords[-1],3)) #returns your last x_n and y_n
                                                        #rounded to three decimal places.

gradient_descent(f_1, 0.01, 1) #call function 1
```

```python
gradient_descent(f_2, 0.01, 1) #call function 2
```

```python
gradient_descent(f_3, 0.01, 1) #call function 3
```

```python
gradient_descent(f_4, 0.01, 1) #call function 4
```
